```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
df = pd.read_csv('Market_Basket_Optimisation.csv',header=None)
```

```python
df
```

```python
df.info()
```

```python
data = []
for i in range(0,7501):
    data.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
assoc_rules = apriori(data, min_support= 0.003,
              min_confidence= 0.2,
              min_lift=3, 
              min_length = 2 , 
              max_length = 4 ) 
assoc_res = list(assoc_rules)
```

```python
len(assoc_res)
```

```python
for i in assoc_res:
        print(i,'\n')
        print('\n\n')
```
